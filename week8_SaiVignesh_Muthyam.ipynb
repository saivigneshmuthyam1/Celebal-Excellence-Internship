````python
# Tool 1: Calculator
def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# Tool 2: Keyword Extractor
def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
# Agent logic implementation
def agent(query: str):
    query_lower = query.lower()

    # check for math words or if the query contains numbers and math symbols
    has_math_symbols = any(char in query for char in ['+', '-', '*', '/', '(', ')'])
    is_math_query = "calculate" in query_lower or "calculator" in query_lower or "compute" in query_lower or (has_math_symbols and any(c.isdigit() for c in query))

    if is_math_query:
        # strip out command words to leave just the math expression
        expr = query_lower
        for word in ["calculate", "calculator", "compute"]:
            expr = expr.replace(word, "")
        expr = expr.strip()

        # fallback if they just typed "calculator" with no numbers
        if not expr:
            return {
                "type": "calculation",
                "result": "Please provide a mathematical expression to evaluate."
            }

        return {
            "type": "calculation",
            "result": calculator(expr)
        }

    elif "keywords" in query_lower:
        # cut out 'keywords from' or just 'keywords' to get the text
        if "keywords from" in query_lower:
            text = query_lower.split("keywords from")[-1].strip()
        else:
            text = query_lower.split("keywords")[-1].strip()
        return {
            "type": "keywords",
            "result": extract_keywords(text)
        }

    else:
        # handling basic general questions
        if "what is machine learning" in query_lower:
            ans = "Machine learning is a type of AI that allows software applications to learn from data and become more accurate."
        else:
            ans = "I am a smart assistant. I can handle math queries, extract keywords, or answer basic questions."

        return {
            "type": "general",
            "result": ans
        }
````

````python
# Interactive loop
while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
